```julia
#la funcion recibe las ecuaciones por parametros
function metodoGAUSS(x,y,z,r,x2,y2,z2,r2,x3,y3,z3,r3)
    #matriz = [x y z r; x2 y2 z2 r2; x3 y3 z3 r3]
    ecuacion_a = [x y z r]
    ecuacion_b = [x2 y2 z2 r2]
    ecuacion_c = [x3 y3 z3 r3]
    sistema_ecc = [x y z r; x2 y2 z2 r2; x3 y3 z3 r3]
    ecuacion_mayor = [0.0 0.0 0.0 0.0]
    
    cont = 1
    cont2 = 1
    cont3 = 2
    cont4 = 1
    cont5 = 1
    cont6 = 1
    y = 0.0
    x = 0.0
    z = 0.0
    multiplicacion = 0.0
    mayor_x = 0.0 
    mayor_y = 0.0 
    
    #Pivoteo para ordenar la matriz
    
    if ecuacion_a[1] >= ecuacion_b[1]
        if ecuacion_a[1] >= ecuacion_c[1]
            if ecuacion_b[1] > ecuacion_c[1]
                matriz = [ecuacion_a; ecuacion_b; ecuacion_c]
            else
                matriz = [ecuacion_a; ecuacion_c; ecuacion_b]
            end
        end
    end
    
    if ecuacion_b[1] >= ecuacion_a[1]
        if ecuacion_b[1] >= ecuacion_c[1]
            if ecuacion_a[1] > ecuacion_c[1]
                matriz = [ecuacion_b; ecuacion_a; ecuacion_c]
            else    
                matriz = [ecuacion_b; ecuacion_c; ecuacion_a]
            end
        end
    end
    
    if ecuacion_c[1] >= ecuacion_a[1]
        if ecuacion_c[1] >= ecuacion_b[1]
            if ecuacion_a[1] > ecuacion_b[1]
                matriz = [ecuacion_c; ecuacion_a; ecuacion_b]
            else
                matriz = [ecuacion_c; ecuacion_b; ecuacion_a]
            end
        end
    end
    
    print("La matriz luego de pivotear es: ", matriz, "\n")
    
    #toma la primera ecuacion y la multiplica por valor de x de la segunda ecuacion partido en x de la primera ecuacion, 
    #luego el resultado se lo resta a la segunda ecuacion
    while cont2 <= 4
        multiplicacion = (matriz[1,cont2]/matriz[1,1])*matriz[2,1]
        matriz[2,cont2] = matriz[2,cont2] - multiplicacion
        cont2 += 1
        
    end 
    
    #toma la primera ecuacion y la multiplica por valor de x de la tercera ecuacion partido en x de la primera ecuacion,
    #luego el resultado se lo resta a la tercera ecuacion
    while cont <= 4
        multiplicacion = (matriz[1,cont]/matriz[1,1])*matriz[3,1]
        matriz[3,cont] = matriz[3,cont] - multiplicacion
        cont += 1
        
    end
    
    #Toma la segunda ecuacion y la multiplica por el valor de y de la tercera ecuacion partido en y de la segunda ecuacion,
    #luego el resultado se lo resta a la terceraecuacion
    while cont3 <= 4
        multiplicacion = (matriz[2,cont3]/matriz[2,2])*matriz[3,2]
        matriz[3,cont3] = matriz[3,cont3] - multiplicacion
        cont3 += 1
        
    end
    
    #calcula los valores de x, y, z por medio de sustitucion. 
    z = matriz[3,4]/matriz[3,3]
    y = (matriz[2,4] - (matriz[2,3]*z))/matriz[2,2]
    x = (matriz[1,4] - matriz[1,3]*z)/matriz[1,1]
    
    #agrupa los valores de x, y, z en un vector y los muestra por pantalla
    vector = [x ; y ; z]
    println("Los valores de las variables x, y, z son respectivamente: ",vector)
    
end

#Sistema de ecuaciones lineales de ejercicio 1 pasado por parametros
metodoGAUSS(0.0, -50.0, 50.0, 784.8, 150.0, 0.0, -100.0, 588.6, -100.0, 150.0, -50.0, 686.7)
```
