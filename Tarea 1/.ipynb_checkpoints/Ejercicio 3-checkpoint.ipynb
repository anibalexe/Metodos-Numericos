```julia
function biseccion()
    a = 0.0
    b = 2.0
    c = 0.0
    evaluacionA = 0.0
    evaluacionB = 0.0
    evaluacionC = 0.0
    errorRelativoAprox = 1.0
    cont = 1
    
    #El ciclo se repite hasta que error es menor al 0.05%
    while errorRelativoAprox >= 0.05
        c = (a + b)/2
        evaluacionA = (a^4)-1-(3*exp(-a))
        evaluacionB = (b^4)-1-(3*exp(-b))
        evaluacionC = (c^4)-1-(3*exp(-c))
        println("Iteracion ",cont)
        println("a es: ",a,", b es: ",b,", c es: ",c,"\nf(a): ",evaluacionA,", f(b): ",evaluacionB,", f(c): ",evaluacionC)
        #Comprueba que valor se reemplazarÃ¡ por la variable c
        if evaluacionA*evaluacionC > 0
            #Calcula el error relativo aproximado porcentual
            errorRelativoAprox = abs((c-a)/c)*100
            a = c
            println("Actualizacion: a = c")
            
        else
            #Calcula el error relativo aproximado porcentual
            errorRelativoAprox = abs((c-b)/c)*100
            b = c
            println("Actualizacion: b = c")

        end
        println("Error relativo aproximado: ",errorRelativoAprox,"%\n")
        cont += 1
        
    end

end

biseccion()
```
