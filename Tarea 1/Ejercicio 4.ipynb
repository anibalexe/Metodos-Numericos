```julia
using Plots

function graficar()
    f(x) = sin((-0.5)*exp(x)+(pi*cos(2*x)))
    Plots.scatter(f)
    #Se tomó dos puntos aproximados cercanos a una raiz
    raices = [2.9, 4.9]
    #Grafica la funcion de 0 a 2pi
    plot([f,zero], 0, 2*pi)
    #Imprime las raices
    scatter!(raices, [0, 0])
    
end

graficar()
```

```julia
function biseccion()
    a = 0.0
    b = 2.0*pi
    c = 0.0
    evaluacionA = 0.0
    evaluacionB = 0.0
    evaluacionC = 0.0
    cont = 1
    
    println("METODO DE LA BISECCIÓN")
    #El ciclo se repite mientras cont sea menor o igual a 14
    while cont <= 14
        c = (a + b)/2
        evaluacionA = sin((-0.5)*exp(a)+(pi*cos(2*a)))
        evaluacionB = sin((-0.5)*exp(b)+(pi*cos(2*b)))
        evaluacionC = sin((-0.5)*exp(c)+(pi*cos(2*c)))
        println("Iteracion ",cont)
        println("a es: ",a,", b es: ",b,", c es: ",c,"\nf(a): ",evaluacionA,", f(b): ",evaluacionB,", f(c): ",evaluacionC)
        
        #Comprueba que valor se reemplazará por la variable c
        if evaluacionA*evaluacionC > 0
            a = c
            println("Actualizacion: a = c\n")
            
        else
            b = c
            println("Actualizacion: b = c\n")
            
        end
        cont += 1
        
    end

end

biseccion()
```
