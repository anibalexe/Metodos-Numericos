```julia
using Plots
function reglaTrapezoidal()
    a = 0.0
    b = 0.8
    n = 4.0
    h = 0.0
    puntos = []
    area = 0.0
    evaluacionA = 0.0
    evaluacionB = 0.0
    cont = 1
    x = 0.0
    
    h = (b-a)/n
    
    #Obtiene los 3 puntos de corte en la curva
    while cont <= (n-1)
        append!(puntos,[a+(cont*h)])
        cont += 1
    end
    
    
    cont = 1
    #Calcula la evaluacion de cada punto en la curva
    while cont <= (n-1)
        x = puntos[cont]
        area += 2*(0.2+(25*x)-(200*x^2)+(675*x^3)-(900*x^4)+(400*x^5))
        cont += 1
    end
    
    #Se suma la evaluacion de a y b al area estimada de la curva 
    evaluacionA = (0.2+(25*a)-(200*a^2)+(675*a^3)-(900*a^4)+(400*a^5))
    evaluacionB = (0.2+(25*b)-(200*b^2)+(675*b^3)-(900*b^4)+(400*b^5))
    area += evaluacionA+evaluacionB
    #Finalmente se obtine el area estimada de la curva entre 0 y 0.8
    area = area*(h/2)
    println("El Ã¡rea estimada bajo la curva es: ",area)
    
    #Grafica la curva 
    f(x) = 0.2+(25*x)-(200*x^2)+(675*x^3)-(900*x^4)+(400*x^5)
    Plots.scatter(f)
    plot([f,zero], 0, 0.8)
    
end

reglaTrapezoidal()
```
