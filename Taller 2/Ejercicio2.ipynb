```julia
function biseccion()
    a = -1.0
    b = 0.0
    m = 0.0
    evaluacionA = 0.0
    evaluacionB = 0.0
    evaluacionM = 0.0
    i = 0
    #El ciclo se repite 100 veces
    while i <= 100
        m = (a + b)/2
        evaluacionA = -12-(21*a)+(18*a^2)-(2.75*a^3)
        evaluacionB = -12-(21*b)+(18*b^2)-(2.75*b^3)
        evaluacionM = -12-(21*m)+(18*m^2)-(2.75*m^3)
        println("f(m) varió :",abs(0-abs(evaluacionM))*100,"% con repeto a y = 0")
        println(" ")
        if evaluacionA*evaluacionM > 0
            a = m
        else
            b = m
        end
        i += 1
    end
    println("Con el metodo de la biseccion la raiz se encuentra cerca del punto: (",m,", ",evaluacionM,")")
    println("")
end

function falsi()
    a = -1.0
    b = 0.0
    x = 0.0
    evaluacionA = 0.0
    evaluacionB = 0.0
    evaluacionX = 0.0
    auxX = 0.0
    auxevaluacionX = 0.0
    error = 1.0
    #El ciclo termina cuando el error es igual a cero
    while error != 0
        auxX = x
        evaluacionA = -12-(21*a)+(18*a^2)-(2.75*a^3)
        evaluacionB = -12-(21*b)+(18*b^2)-(2.75*b^3)
        auxevaluacionX = evaluacionX
        x = b-((evaluacionB*(a-b))/(evaluacionA-evaluacionB))
        evaluacionX = -12-(21*x)+(18*x^2)-(2.75*x^3)
        #Error relativo
        error = abs(evaluacionX-auxevaluacionX)/evaluacionX
        if evaluacionA*evaluacionX > 0
            a = x
        else
            b = x
        end
    end
    println("Con el metodo falsi la raiz se encuentra cerca del punto: (",x,", ",evaluacionX,")")
end

biseccion()

falsi()
```
