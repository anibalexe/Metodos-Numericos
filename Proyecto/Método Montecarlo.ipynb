```julia
using Plots
function metodoMontecarlo()
    #Funcion a la que se le buscara el rea
    f(i) = (200*i^sin(i))-(i^2)
    #Atributos del modelo matematico
    totalPuntos = 2000
    puntosExito = 0
    cotaSuperior = 1550
    a = 0
    b = 10
    x = 0.0
    y = 0.0
    area = 0.0
    xRand = []
    yRand = []
    
    #Ciclo que se repite incorporando todo los puntos en la curva
    cont = 1
    while cont <= totalPuntos
        #Genera puntos random en las coordenadas x e y pertenecientes a los limites
        x = a+rand()*(b-a)
        y = rand()*cotaSuperior
        #Almacena los puntos generados
        append!(xRand, [x])
        append!(yRand, [y])
        #Calcula el punto en y real al evaluar el punto x random
        yReal = (200*x^sin(x))-(x^2)
        #Verifica si el punto random en la coordenada y cae dentro de la curva y lo cuenta como punto de exito
        if yReal >= y
            puntosExito +=1 
            
        end
        cont += 1
        
    end
    
    #Calcula el area bajo la curva
    area = (puntosExito/(totalPuntos))*(b-a)*cotaSuperior
    
    println("Total de puntos dados: ",totalPuntos,", Puntos de exito: ",puntosExito,", Area calculada: ",area)
    #Imprime el grafico con el total de puntos generados
    Plots.scatter(f)
    plot([f,zero], a, b)
    scatter!(xRand,yRand)
    
end

metodoMontecarlo()
```
