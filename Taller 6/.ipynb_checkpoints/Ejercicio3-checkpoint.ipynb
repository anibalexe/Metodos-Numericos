```julia
function multiple()
    vectorX1 = [0,1,1,2,2,3,3,4,4]
    vectorX2 = [0,1,2,1,2,1,2,1,2]
    vectorY = [15.1,17.9,12.7,25.6,20.5,35.1,29.7,45.4,40.2]
    sumatoriaX1 = 0.0
    sumatoriaX2 = 0.0
    sumatoriaX1cuadrado = 0.0
    sumatoriaX2cuadrado = 0.0
    sumatoriaY = 0.0
    sumatoriaX1X2 = 0.0
    sumatoriaX1Y = 0.0
    sumatoriaX2Y = 0.0
    n = 8
    cont = 1
    cont1 = 1
    cont2 = 1
    cont3 = 1
    sr = 0.0
    st = 0.0
    rcuadrado = 0.0
    promedioY = 0.0
    
    for i in vectorX1
        sumatoriaX1 += i
        sumatoriaX1cuadrado += i^2
    end
    
    for i in vectorX2
        sumatoriaX2 += i
        sumatoriaX2cuadrado += i^2
    end
    
    for i in vectorY
        sumatoriaY += i
    end
    
    while cont1 <= 8
        sumatoriaX1X2 += vectorX1[cont1]*vectorX2[cont1]
        sumatoriaX1Y += vectorX1[cont1]*vectorY[cont1]
        sumatoriaX2Y += vectorX2[cont1]*vectorY[cont1]
        cont1 += 1
    end
    
    matriz = [n sumatoriaX1 sumatoriaX2 sumatoriaY; sumatoriaX1 sumatoriaX1cuadrado sumatoriaX1X2 sumatoriaX1Y; sumatoriaX2 sumatoriaX1X2 sumatoriaX2cuadrado sumatoriaX2Y]
    
        #toma la primera ecuacion y la multiplica por valor de x de la segunda ecuacion partido en x de la primera ecuacion, 
    #luego el resultado se lo resta a la segunda ecuacion
    while cont2 <= 4
        multiplicacion = (matriz[1,cont2]/matriz[1,1])*matriz[2,1]
        matriz[2,cont2] = matriz[2,cont2] - multiplicacion
        cont2 += 1
        
    end 
    
    #toma la primera ecuacion y la multiplica por valor de x de la tercera ecuacion partido en x de la primera ecuacion,
    #luego el resultado se lo resta a la tercera ecuacion
    while cont <= 4
        multiplicacion = (matriz[1,cont]/matriz[1,1])*matriz[3,1]
        matriz[3,cont] = matriz[3,cont] - multiplicacion
        cont += 1
        
    end
    
    #Toma la segunda ecuacion y la multiplica por el valor de y de la tercera ecuacion partido en y de la segunda ecuacion,
    #luego el resultado se lo resta a la terceraecuacion
    while cont3 <= 4
        multiplicacion = (matriz[2,cont3]/matriz[2,2])*matriz[3,2]
        matriz[3,cont3] = matriz[3,cont3] - multiplicacion
        cont3 += 1
        
    end
    
    #calcula los valores de x, y, z por medio de sustitucion. 
    a2 = matriz[3,4]/matriz[3,3]
    a1 = (matriz[2,4] - (matriz[2,3]*a2))/matriz[2,2]
    a0 = (matriz[1,4] - matriz[1,3]*a2)/matriz[1,1]
    
    #agrupa los valores de x, y, z en un vector y los muestra por pantalla
    vector = [a0 ; a1 ; a2]
    println("Los valores de las variables a0, a1, a2 son respectivamente: ",vector)
    
    promedioY = sumatoriaY/n
    
    while cont <= 8
        sr = sr + (vectorY[cont] - a0 - a1*vectorX1[cont] - a2*vectorX2[cont])^2
        st = st + (vectorY[cont] - promedioY)^2
        cont += 1
    end
    rcuadrado = (st - sr)/st
    println("st: ",st,", sr: ",sr,", r al cuadrado: ",rcuadrado)
end

multiple()
```
